```python
# Капушев Сергей, Вариант 2

import pandas as pd   # основная библеотека для работы


# Ссылка на csv файл с данными
url = "https://gist.githubusercontent.com/Hozard0/5733edf12ed98d63af02bc7bfb654a7e/raw/3ae5e6776c1cb98875b4a416b75254000a0a7f33/%25D1%2581%25D0%25BE%25D1%2582%25D1%2580%25D1%2583%25D0%25B4%25D0%25BD%25D0%25B8%25D0%25BA%25D0%25B8_%25D1%2582%25D0%25B0%25D0%25B1%25D0%25BB%25D0%25B8%25D1%2586%25D0%25B0_%25D1%2581_%25D0%25B4%25D0%25B0%25D1%2582%25D0%25B0%25D0%25BC%25D0%25B8.csv"

# Передадим csv файл в DataFrame
df = pd.read_csv(url)

# Узнаем настоящее время (по Томскому времени)
current_datetime = pd.Timestamp.now()
current_datetime = current_datetime+pd.Timedelta(hours = 7)
print(f"Дата: {current_datetime}")






```

```python
# Преобразуем str в date, и находим стаж работы в годах
date_series = pd.to_datetime(df['Дата приема на работу'])
df['Стаж'] = (current_datetime - date_series).dt.days // 365

# Фильтрация стажа
def stage(asg):
  if asg < 3:
    return 'Мало'
  elif (asg >= 3) & (asg <= 7):
    return 'Средний'
  elif asg > 7:
    return 'Большой'
df['Категория стажа'] = df['Стаж'].apply(stage)

# Удаление ненужного столбца
df.drop('Стаж', axis = 1, inplace = True)
df
```
